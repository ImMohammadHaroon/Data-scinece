```python
#  1. Create Series

s = pd.Series([10, 20, 30, 40, 50], index=['a', 'b', 'c', 'd', 'e'])
s
```

```python
# Basic Methods
print("sum:", s.sum())
print("mean:", s.mean())
print("max:", s.max())
print("min:", s.min())
print("describe:\n", s.describe())
```

```python
#  2. Indexing in Series

# Label indexing
print("Label indexing s['c']:", s['c'])

# Integer location indexing
print("Integer indexing s[2]:", s[2])

# Slicing
print("Slicing s['b':'d']:\n", s['b':'d'])

# Boolean indexing
print("Boolean indexing s[s > 25]:\n", s[s > 25])

# Fancy indexing
print("Fancy indexing s[['a', 'e', 'c']]:\n", s[['a', 'e', 'c']], "\n")
```

```python
#  3. Create DataFrame
data = {
    'Name': ['Ali', 'Sara', 'John', 'Mina', 'Tom'],
    'Age': [23, 25, 30, 22, 28],
    'Marks': [88, 92, 79, 95, 85],
    'City': ['Lahore', 'Karachi', 'London', 'Dubai', 'Paris']
}
df = pd.DataFrame(data, index=['a', 'b', 'c', 'd', 'e'])
print(df)
```

```python
df.T
```

```python
# 4. DataFrame Basic Methods
print("Shape:", df.shape)
print("Info:")
print(df.info(), "\n")
print("Describe:\n", df.describe(), "\n")
print("Head:\n", df.head(3))
print("Tail:\n", df.tail(2), "\n")
```

```python
# 5. Indexing and Selection in DataFrame
# Column selection
print("Select one column (Age):\n", df['Age'], "\n")

# Multiple columns
print("Select multiple columns (Name & City):\n", df[['Name', 'City']], "\n")

# Row selection by label (loc)
print("Row selection by label (loc['c']):\n", df.loc['c'], "\n")

# Row selection by position (iloc)
print("Row selection by position (iloc[2]):\n", df.iloc[2], "\n")
```

```python
# Row + Column selection using loc
print("loc['b', 'Marks']:", df.loc['b', 'Marks'])
print("loc[['a','d'], ['Name','City']]:\n", df.loc[['a', 'd'], ['Name', 'City']], "\n")

# Row + Column selection using iloc
print("iloc[1:4, 1:3]:\n", df.iloc[1:4, 1:3], "\n")

# Slicing rows
print("Slice rows df[1:4]:\n", df[1:4], "\n")

# Boolean indexing
print("Boolean indexing df[df['Marks'] > 85]:\n", df[df['Marks'] > 85], "\n")

# Conditional filtering with multiple conditions
print("Students older than 23 and Marks > 85:\n", df[(df['Age'] > 23) & (df['Marks'] > 85)], "\n")
```

```python

# Add new column
df['Grade'] = ['B+', 'A', 'B', 'A+', 'B']
print("Added new column 'Grade':\n", df, "\n")

# Modify a column
df['Marks'] = df['Marks'] + 5
print("Modified 'Marks' column (+5):\n", df, "\n")

# Delete a column
df2 = df.drop('City', axis=1)
print("Deleted 'City' column:\n", df2, "\n")

# Sort values
print("Sort by 'Age':\n", df.sort_values(by='Age'), "\n")

# Sort by index
print("Sort by index descending:\n", df.sort_index(ascending=False), "\n")
```

```python
#  7. Handling Missing Data
df_missing = df.copy()
df_missing.loc['f'] = [np.nan, np.nan, np.nan, np.nan, np.nan]
print("With Missing Row:\n", df_missing, "\n")

print("Check for NaN:\n", df_missing.isna(), "\n")
print("Drop NaN rows:\n", df_missing.dropna(), "\n")
print("Fill NaN with default values:\n", df_missing.fillna({'Name': 'Unknown', 'Age': 0, 'Marks': 0}), "\n")
```

```python
#  8. Statistical & Mathematical Operations

print("Mean Marks:", df['Marks'].mean())
print("Max Age:", df['Age'].max())
print("Correlation:\n", df.corr(numeric_only=True))
print("Apply function (square of marks):\n", df['Marks'].apply(lambda x: x**2), "\n")
```

```python
# 10. GroupBy Example

data2 = {
    'Dept': ['CS', 'CS', 'Math', 'Math', 'Physics'],
    'Score': [90, 85, 88, 70, 95]
}
df_group = pd.DataFrame(data2)
print("DataFrame for GroupBy:\n", df_group, "\n")

print("Group by Dept, Mean Scores:\n", df_group.groupby('Dept')['Score'].mean(), "\n")

```

```python

# 1. Create Index objects

idx1 = pd.Index(['a', 'b', 'c', 'd', 'e'])
idx2 = pd.Index(['c', 'd', 'e', 'f', 'g'])
print("Index 1:", idx1)
print("Index 2:", idx2, "\n")

# 2. append() — Concatenate Index objects
appended = idx1.append(idx2)
print(appended, "\n")

# 3. difference() — Set difference

diff = idx1.difference(idx2)
print(diff, "\n")

# 4. intersection() — Common elements

inter = idx1.intersection(idx2)
print(inter, "\n")

# 5. union() — Combine all unique elements
uni = idx1.union(idx2)
print(uni, "\n")

# 6. isin() — Check if elements of idx1 are in idx2
isin_mask = idx1.isin(idx2)
print(isin_mask)
print("Elements in idx1 also in idx2:", idx1[isin_mask], "\n")

# 7. delete() — Delete element by position
deleted = idx1.delete(2)
print(deleted, "\n")

#  8. drop() — Delete element(s) by label
dropped = idx1.drop(['a', 'd'])
print(dropped, "\n")

#  9. insert() — Insert new label at given position
inserted = idx1.insert(2, 'z')
print(inserted, "\n")

#  10. is_monotonic — True if sorted ascending
print("idx1:", idx1.is_monotonic_increasing)
idx_sorted = pd.Index(['a', 'b', 'c', 'd'])
print("Sorted Index:", idx_sorted.is_monotonic_increasing, "\n")

# 11. is_unique — True if all values are unique
idx_dup = pd.Index(['a', 'b', 'b', 'c'])
print("idx1 unique?:", idx1.is_unique)
print("idx_dup unique?:", idx_dup.is_unique, "\n")

# 12. unique() — Return array of unique values

print("Unique values in idx_dup:", idx_dup.unique(), "\n")

# 12. unique() — Return array of unique values

# BONUS: Show in context of DataFrame index
df = pd.DataFrame({
    "Name": ["Ali", "Sara", "John", "Mina"],
    "Marks": [88, 92, 85, 90]
}, index=['a', 'b', 'c', 'd'])

print("Original DataFrame:\n", df, "\n")

# Demonstrate .index methods on DataFrame
print("DataFrame Index union with another Index ['d','e','f']:\n", df.index.union(pd.Index(['d','e','f'])))
print("DataFrame Index difference with ['b','c']:\n", df.index.difference(pd.Index(['b','c'])))
print("DataFrame Index is_unique:", df.index.is_unique)
```
