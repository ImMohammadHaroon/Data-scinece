```python
import numpy as np

# 1ï¸âƒ£ permutation()
arr1 = np.array([1, 2, 3, 4, 5])
perm = np.random.permutation(arr1)
# print("1ï¸âƒ£ permutation() - Random permutation of array:")
print("Original:", arr1)
print("Permutation:", perm, "\n")

# 2ï¸âƒ£ shuffle()
arr2 = np.array([10, 20, 30, 40, 50])
np.random.shuffle(arr2)
# print("2ï¸âƒ£ shuffle() - Shuffle array in-place:")
print("Shuffled:", arr2, "\n")

# 3ï¸âƒ£ uniform()
uniform_arr = np.random.uniform(low=0.0, high=10.0, size=(3, 3))
# print("3ï¸âƒ£ uniform() - Uniform distribution [0,10):")
print(uniform_arr, "\n")

# 4ï¸âƒ£ integers()
int_arr = np.random.randint(low=0, high=100, size=(3, 3))
# print("4ï¸âƒ£ integers() - Random integers from 0â€“99:")
print(int_arr, "\n")

# 5ï¸âƒ£ standard_normal()
stdnorm_arr = np.random.standard_normal(size=(3, 3))
# print("5ï¸âƒ£ standard_normal() - Normal distribution (mean=0, std=1):")
print(stdnorm_arr, "\n")

# 6ï¸âƒ£ binomial()
binom_arr = np.random.binomial(n=10, p=0.5, size=5)
# print("6ï¸âƒ£ binomial() - Binomial distribution (n=10, p=0.5):")
print(binom_arr, "\n")

# 7ï¸âƒ£ normal()
normal_arr = np.random.normal(loc=50, scale=10, size=(3, 3))
# print("7ï¸âƒ£ normal() - Gaussian distribution (mean=50, std=10):")
print(normal_arr, "\n")

# 8ï¸âƒ£ beta()
beta_arr = np.random.beta(a=2.0, b=5.0, size=5)
# print("8ï¸âƒ£ beta() - Beta distribution (a=2, b=5):")
print(beta_arr, "\n")

# 9ï¸âƒ£ chisquare()
chi_arr = np.random.chisquare(df=2, size=5)
# print("9ï¸âƒ£ chisquare() - Chi-square distribution (df=2):")
print(chi_arr, "\n")

# ðŸ”Ÿ gamma()
gamma_arr = np.random.gamma(shape=2.0, scale=1.0, size=5)
# print("ðŸ”Ÿ gamma() - Gamma distribution (shape=2, scale=1):")
print(gamma_arr, "\n")

# BONUS: uniform() [0, 1)
uni_default = np.random.uniform(size=(3, 3))
# print("ðŸ”¹ BONUS uniform() - Uniform [0,1) default:")
print(uni_default, "\n")
```

```python
import numpy as np

# Base array
arr = np.array([-3.5, -0.2, 0, 2.3, 5.7])
arr2 = np.array([2, 4, 0.5, -1, -3])

print("Original Array:\n", arr)
print("Second Array:\n", arr2)
print("="*70)

# -------------------- Unary Universal Functions --------------------

# print("\nðŸ”¹ abs / fabs:")
print(np.abs(arr))
print(np.fabs(arr))   # fabs only works on real numbers

# print("\nðŸ”¹ sqrt (square root):")
print(np.sqrt(np.abs(arr)))   # abs to avoid sqrt of negatives

# print("\nðŸ”¹ square (xÂ²):")
print(np.square(arr))

# print("\nðŸ”¹ exp (e^x):")
print(np.exp(arr))

# print("\nðŸ”¹ log / log10 / log2 / log1p:")
pos = np.abs(arr) + 1e-3  # avoid log(0)
print("log:", np.log(pos))
print("log10:", np.log10(pos))
print("log2:", np.log2(pos))
print("log1p:", np.log1p(pos))

# print("\nðŸ”¹ sign (1, 0, -1):")
print(np.sign(arr))

# print("\nðŸ”¹ ceil:")
print(np.ceil(arr))

# print("\nðŸ”¹ floor:")
print(np.floor(arr))

# print("\nðŸ”¹ rint (round to nearest int):")
print(np.rint(arr))

# print("\nðŸ”¹ modf (fractional, integral parts):")
frac, integ = np.modf(arr)
print("Fractional:", frac)
print("Integral:", integ)

# print("\nðŸ”¹ isnan (is NaN):")
nan_arr = np.array([1, np.nan, np.inf, -np.inf])
print(np.isnan(nan_arr))

# print("\nðŸ”¹ isfinite / isinf:")
print("isfinite:", np.isfinite(nan_arr))
print("isinf:", np.isinf(nan_arr))

# print("\nðŸ”¹ Trigonometric functions (cos, sin, tan):")
print("cos:", np.cos(arr))
print("sin:", np.sin(arr))
print("tan:", np.tan(arr))

# print("\nðŸ”¹ Hyperbolic trig functions (cosh, sinh, tanh):")
print("cosh:", np.cosh(arr))
print("sinh:", np.sinh(arr))
print("tanh:", np.tanh(arr))

#   print("\nðŸ”¹ Inverse trig functions (arccos, arcsin, arctan):")
limited = np.clip(arr / 6, -1, 1)   # keep within valid domain [-1,1]
print("arccos:", np.arccos(limited))
print("arcsin:", np.arcsin(limited))
print("arctan:", np.arctan(limited))

#   print("\nðŸ”¹ Inverse hyperbolic trig functions (arccosh, arcsinh, arctanh):")
pos_vals = np.abs(arr) + 1
frac_vals = np.clip(arr / 6, -0.9, 0.9)
print("arccosh:", np.arccosh(pos_vals))
print("arcsinh:", np.arcsinh(arr))
print("arctanh:", np.arctanh(frac_vals))

# print("\nðŸ”¹ logical_not (NOT):")
bool_arr = np.array([True, False, True, False])
print(np.logical_not(bool_arr))

print("="*70)
# -------------------- Binary Universal Functions --------------------

# print("\nðŸ”¹ add:")
print(np.add(arr, arr2))

# print("\nðŸ”¹ subtract:")
print(np.subtract(arr, arr2))

# print("\nðŸ”¹ multiply:")
print(np.multiply(arr, arr2))

# print("\nðŸ”¹ divide:")
print(np.divide(arr, arr2))

# print("\nðŸ”¹ floor_divide:")
print(np.floor_divide(arr, arr2))

# print("\nðŸ”¹ power (x^y):")
print(np.power(arr, 2))

# print("\nðŸ”¹ maximum / fmax:")
nan_mix = np.array([np.nan, 2, 5])
print("maximum:", np.maximum(nan_mix, 3))
print("fmax:", np.fmax(nan_mix, 3))   # ignores NaN

# print("\nðŸ”¹ minimum / fmin:")
print("minimum:", np.minimum(nan_mix, 3))
# print("\nðŸ”¹ mod (remainder):")
print(np.mod(arr, arr2))

# print("\nðŸ”¹ copysign (copy sign of arr2 to arr):")
print(np.copysign(arr, arr2))

# print("\nðŸ”¹ Comparison operators (greater, less_equal, equal, not_equal):")
print("greater:", np.greater(arr, arr2))
print("greater_equal:", np.greater_equal(arr, arr2))
print("less:", np.less(arr, arr2))
print("less_equal:", np.less_equal(arr, arr2))
print("equal:", np.equal(arr, arr2))
print("not_equal:", np.not_equal(arr, arr2))

# print("\nðŸ”¹ logical operations (and, or, xor):")
bool1 = np.array([True, False, True])
bool2 = np.array([False, False, True])
print("logical_and:", np.logical_and(bool1, bool2))
print("logical_or:", np.logical_or(bool1, bool2))
print("logical_xor:", np.logical_xor(bool1, bool2))
```
